```python
"""
year = 2021
month = 03
day = 25
"""

# datetime menjadi standard library yang bisa membantu kasus ini
# modul spesifik yang digunakan adalah date
from datetime import date 

# variabel year, month, day yang akan digunakan
year = 2021
month = 3
day = 25

date = date(year, month, day) # variabel yang awalnya integer akan dikonversi ke tipe data datetime
iso_date = date.isoformat() # penggunaan func isoformat digunakan untuk mengkonversi dari date object ke string

print(date, type(date))
print(iso_date, type(iso_date))
```

from datetime import date 


```python
from datetime import date 

# Untuk mendapatkan tanggal dari hari ini, kita bisa menggunakan func today 
today = date.today() 
part_date = date(2021, 3, 25)

diff = today - part_date

# func strftime dengan paramater %A akan mengembalikan nama lengkap dari hari tanggal yang diinginkan 
day_name = part_date.strftime("%A") 

print(today, part_date, day_name, diff, sep="\n")
```

## Exercise 5

```python
import pandas as pd
import datetime as dt

start_date = dt.date.today()

date_list = pd.date_range(start_date, periods=10, freq='W')

df = pd.DataFrame({
    'Date': date_list,
    'Day_Name': date_list.strftime('%A')
})

print(f"--- What Date is Break Day ---")
print(df)
```

## Exercise 6

```python
import pandas as pd
import datetime as dt
import QOTD

start_date = dt.date.today()

date_list = pd.date_range(start_date, periods=10, freq='W')

message = QOTD.get_random_messages(len(date_list))

df = pd.DataFrame({
    'Date': date_list,
    'Day_Name': date_list.strftime('%A'),
    'Messages': message
})

print("\t--- What Date is Break Day ---\n")
print(df)
```
