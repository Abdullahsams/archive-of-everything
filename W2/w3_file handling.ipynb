# Latihan 1

```python
import os

def list_files():
    print("\nDaftar file di direktori saat ini:")
    files = [f for f in os.listdir() if f.endswith('.txt')] # mengecek daftar file yang tersedia di direktori menggunakan os.listdir(), dengan kondisi file berekstensi *.txt
    if not files:
        print("- Tidak ada file .txt ditemukan.")
    for file in files:
        print(f"- {file}")

def write_to_file():
    filename = input("Masukkan nama file (contoh: data.txt): ")
    content = input("Masukkan teks yang ingin ditulis: ")
    f = open(filename, "a") # membuka file dengan mode a untuk menulis, dan menambahkan ke baris terakhir jika duplikat
    f.write(content + "\n")
    f.close() # menutup data, memastikan tidak terbuka terus-menerus di memory
    print("Teks berhasil ditambahkan!")

def read_from_file():
    filename = input("Masukkan nama file yang ingin dibaca: ")
    if os.path.isfile(filename):
        f = open(filename, "r") # membuka file dengan mode r untuk membaca file saja
        print("\n--- Isi File ---")
        print(f.read()) # membaca seluruh isi file, membaca sejumlah byte pertama secara default 
        f.close() # menutup data, memastikan tidak terbuka terus-menerus di memory
        print("----------------")
    else:
        print("File tidak ditemukan!")

def main():
    while True:
        print("\n=== HANDLING TEXT FILE ===")
        print("1. Lihat Daftar File")
        print("2. Tulis/Tambah Teks ke File")
        print("3. Baca Isi File")
        print("4. Keluar")
        pilihan = input("Pilih menu (1-4): ")

        if pilihan == '1':
            list_files()
        elif pilihan == '2':
            write_to_file()
        elif pilihan == '3':
            read_from_file()
        elif pilihan == '4':
            print("Program selesai.")
            break
        else:
            print("Pilihan tidak valid!")

main()
```

# Latihan 2

```python
import os

def list_files():
    print("\nDaftar file di direktori saat ini:")
    files = [f for f in os.listdir() if f.endswith('.txt')] # mengecek daftar file yang tersedia di direktori menggunakan os.listdir(), dengan kondisi file berekstensi *.txt
    if not files:
        print("- Tidak ada file .txt ditemukan.")
    for file in files:
        print(f"- {file}")

def write_to_file():
    filename = input("Masukkan nama file (contoh: data.txt): ")
    content = input("Masukkan teks yang ingin ditulis: ")
    with open(filename, "a") as f: # penutupan file dilakukan secara otomatis menggunakan 'with' tanpa menutupnya secara manual meggunakan close() func
        f.write(content + "\n")
    print("Teks berhasil ditambahkan!")

def read_from_file():
    filename = input("Masukkan nama file yang ingin dibaca: ")
    if os.path.isfile(filename):
        with open(filename, "r") as f:
            print("\n--- Isi File ---")
            print(f.read())
        print("----------------")
    else:
        print("File tidak ditemukan!")

def main():
    while True:
        print("\n=== HANDLING TEXT FILE ===")
        print("1. Lihat Daftar File")
        print("2. Tulis/Tambah Teks ke File")
        print("3. Baca Isi File")
        print("4. Keluar")
        pilihan = input("Pilih menu (1-4): ")

        if pilihan == '1':
            list_files()
        elif pilihan == '2':
            write_to_file()
        elif pilihan == '3':
            read_from_file()
        elif pilihan == '4':
            print("Program selesai!")
            break
        else:
            print("Pilihan tidak valid!")

main()
```

# Latihan 3

```python
import os

filename = "name.txt"

while True:
    nama = input("masukkan nama: ")
    
    # membaca semua isi file untuk pengecekan
    f = open(filename, "r")
    daftar_nama = f.read().splitlines() # memecah multi-line string menjadi list dengan splitlines() method
    f.close()
    
    # mengecek apakah nama sudah ada dalam daftar
    if nama in daftar_nama:
        print("nama sudah ada pada daftar")
    else:
        # menambahkan nama ke file jika belum ada
        f = open(filename, "a")
        f.write(nama + "\n")
        f.close()
    
    ulang = input("ulangi masukkan nama (y/t)?")
    if ulang.lower() == 't':
        break
```

# Latihan 4

```python
import os

while True:
    nama_file = input("masukkan nama file: ")
    # menambahkan ekstensi .txt untuk pengecekan dan pembuatan
    filename = nama_file + ".txt"
    
    # mengambil daftar file di working dir
    isi_direktori = os.listdir()
    
    # mengecek apakah file sudah ada
    if filename in isi_direktori:
        print("file sudah ada pada direktori tersebut")
    else:
        # membuat file baru
        f = open(filename, "w")
        f.close()
        print(f"{filename} berhasil dibuat")
    
    ulang = input("ulangi membuat file baru (y/t)?")
    if ulang.lower() == 't':
        break
```
