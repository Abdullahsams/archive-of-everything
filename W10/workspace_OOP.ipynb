## Latihan 1

```python
class DataRecord:
  tax_rate = 0.1

  def __init__(self, product_name, quantity, price):
    self.product_name = product_name
    self.quantity = quantity
    self.price = price

  def calculate_total_price(self):
    return self.quantity * self.price

  def calculate_total_price_with_tax(self):
    total = self.calculate_total_price()
    return total + (total * DataRecord.tax_rate)

  def __str__(self):
    return f"{self.product_name} terjual sejumlah {self.quantity} harga {self.price:,}/pc"
```

```python
data_record_1 = DataRecord("Buku tulis", 11, 25500)
data_record_2 = DataRecord("Sticky notes", 5, 8000)
data_record_3 = DataRecord("Rautan pensil", 3, 4000)
```

```python
print(data_record_1)
print(data_record_2)
print(data_record_3)
```

```python
print(data_record_1.calculate_total_price())
print(data_record_2.calculate_total_price())
print(data_record_3.calculate_total_price())
```

```python
print(data_record_1.calculate_total_price_with_tax())
print(data_record_2.calculate_total_price_with_tax())
print(f"{data_record_3.calculate_total_price_with_tax():,.0f}")
```

## Latihan 2

```python
import re

class Dataset:
  def __init__(self, name):
    self.name = name
    self.records = []

  def load_data_from_file(self, filename):
    try:
      with open(filename, "r") as file:
        lines = file.readlines()

      added_record_count = 0 # inisialisasi var yang menampung jumlah data record yang ditambahkan
      for line in lines:
        match = re.match(r'(.+)\s\((\d+)\)\s\-\s(\d+)', line) # Untuk mengidentifikasi nama produk, kuantitas, dan harga. Terbentuk 3 group.

        if match: # Dibentuk masing-masing variabel untuk tiap objek yang ditampung pada group
          product_name = match.group(1)
          quantity = int(match.group(2))
          price = int(match.group(3))

          record = DataRecord(product_name, quantity, price) # Inisialisasi objek baru dari kelas DataRecord
          self.add_data(record)
          added_record_count += 1

      if added_record_count > 1:
        print(f"{added_record_count} data ditambahkan ke {self.name}")
      return added_record_count # Mengembalikan nilai jumlah data record

    except FileNotFoundError:
      print(f"File {filename} tidak ditemukan\n")
      return None # Mengembalikan nilai None jika file tidak ada pada directory

  def add_data(self, record: DataRecord):
    if isinstance(record, DataRecord):
      self.records.append(record)
      print(f"{record.product_name} ditambahkan ke {self.name}")
      return True
    else:
      print(f"{record} bukan objek dari DataRecord\n")
      return False

  def display_data(self):
    print(f"{self.name}:")
    for number, record in enumerate(self.records, start=1):
      print(f"{number}. {record}")
```

```python
dataset_alpha = Dataset("dataset_alpha")

dataset_alpha.add_data(data_record_1)
dataset_alpha.add_data(data_record_2)
dataset_alpha.add_data(data_record_3)
dataset_alpha.add_data("Correction tape")

dataset_alpha.display_data()
```

```python
dataset_beta = Dataset("dataset_beta")

dataset_beta.load_data_from_file("data_penjualan.txt")
dataset_beta.load_data_from_file("penjualan.txt")

dataset_beta.display_data()
```

# Latihan 3

```python
class DataAnalyzer:
  def __init__(self, dataset):
    self.dataset = dataset # "dataset" ini menjadi attribute dari setiap instance, didefinisikan pada special method ini

  def calculate_total_sales(self, include_tax=False):
    if len(self.dataset.records) == 0: # Dilakukan pengecekan apabila dataset dari instance terkait kosong
      return None # Mengembalikan nilai None jika pengecekan di atas terpenuhi
    else:
      total_sales = 0
      for record in self.dataset.records: # Unpack data dari records list yang terdapat pada dataset terkait
        if include_tax: # Pengecekan paramater include_tax bernilai true
          total_sales += record.calculate_total_price_with_tax() # Memanggil method dari kelas DataRecord
        else: # Apabila param include_tax bernilai False
          total_sales += record.calculate_total_price() # Memanggil method dari kelas DataRecord
      return total_sales # Mengembalikan jumlah penjualan yang sudah dilakukan proses perhitungan

  def find_best_selling(self):
    if len(self.dataset.records) == 0:
      return None
    else:
      product_total_qty = {} # Inisialisasi var berupa dictionary untuk menampung best_product (key) dan max_qty (value)

      best_product = None # inisialiasi var yang menampung produk terbaik
      for record in self.dataset.records: # Unpack data dari records list dari dataset terkait
        if record.product_name in product_total_qty: # Melakukan pengecekan apakah nama produk terdapat di dictionary sebagai key
          product_total_qty[record.product_name] += record.quantity # menambahkan kuantitas produk ke total kuantitas yang sudah dicatat untuk produk tersebut"
        else:
          product_total_qty[record.product_name] = record.quantity

      max_qty = 0
      for product_name, qty in product_total_qty.items(): # Unpack nama produk dan kuantitas
        # Melakukan pengecekan produk mana yang memiliki kuantitas penjualan terbanyak, data akan terus diperbarui selama iterasi
        if qty > max_qty:
          max_qty = qty
          best_product = product_name

      return best_product # Mengembalikan nama produk dengan kuantitas penjualan terbesar

  def calculate_average_price(self):
    if len(self.dataset.records) == 0:
      return None
    else:
      total_price_sum = 0
      for record in self.dataset.records:
        total_price_sum += record.price
      return total_price_sum / len(self.dataset.records) # Menghitung rata-rata dari total harga (record.price) dibagi dengan banyaknya data (panjang list records)

  # Tambahan method untuk menampilkan hasil analisis
  def display_analysis(self):
    print(f"Analisis data dari {self.dataset.name}")

    total_sales_no_tax = self.calculate_total_sales()
    total_sales_with_tax = self.calculate_total_sales(include_tax=True)
    best_selling_product = self.find_best_selling()
    average_price = self.calculate_average_price()

    print(f"Total penjualan tanpa pajak: {total_sales_no_tax:,.0f}" if total_sales_no_tax is not None else "Total penjualan tanpa pajak: N/A")
    print(f"Total penjualan dengan pajak: {total_sales_with_tax:,.0f}" if total_sales_with_tax is not None else "Total penjualan dengan pajak: N/A")
    print(f"Produk dengan penjualan tertinggi: {best_selling_product}" if best_selling_product is not None else "Produk dengan penjualan tertinggi: N/A")
    print(f"Rata-rata harga produk: {average_price:,.0f}" if average_price is not None else "Rata-rata harga produk: N/A")
```

```python
analyzer_alpha = DataAnalyzer(dataset_alpha)

DataAnalyzer.display_analysis(analyzer_alpha)
```

```python
analyzer_beta = DataAnalyzer(dataset_beta)

DataAnalyzer.display_analysis(analyzer_beta)
```

```python
dataset_cahyadi = Dataset("dataset_cahyadi")

analyzer_cahyadi = DataAnalyzer(dataset_cahyadi)

DataAnalyzer.display_analysis(analyzer_cahyadi)
```
