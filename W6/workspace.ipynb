```python
passwords = [
    "Kopi_Hitam#2024",     # 15 chars, mixed types
    "S3marang!City@",      # 14 chars, mixed types
    "Batik_P3kalongan!",   # 17 chars, mixed types
    "J4karta_M4ju2026!",   # 18 chars, mixed types
    "P4ssw0rd_Str0ng!",    # 16 chars, mixed types
    "short1!Aa",           # Too short (under 12 chars)
    "hanyahurufkecil!!",   # No uppercase, no numbers
    "HANYAHURUFBESAR123",  # No lowercase, no special chars
    "123456789012345",     # Only numbers
    "StrongPassword123"    # No special characters
]

# DUMMY PASSWORDS AI GENERATED
```

```python
import re
pattern_pwd = r'(?=.*[A-Z])(?=.*[a-z])(?=.*[0-9])(?=.*[!@#$&*]).{12,}'

print('=== STRONG PASSWORD VALIDATOR ===\n')

for password in passwords:
    if re.search(pattern_pwd, password):
        print(f'{password:<20}: Valid')
    else:
        print(f'{password:<20}: Tidak Valid')
```

```python
import re
pattern_pwd = r'^(?=.*[A-Z])(?=.*[a-z])(?=.*[0-9])(?=.*[!@#$&*]).{12,}$'

print('=== STRONG PASSWORD VALIDATOR ===\n')

for password in passwords:
    if re.fullmatch(pattern_pwd, password):
        print(f'{password:<20}: Valid')
    else:
        print(f'{password:<20}: Tidak Valid')
```

## Exercise 4

```python
import re

with open('berita.txt', 'r') as file:
    text = file.readline()
    print(text)
```

Input:
1. Prof. Dr. Ir. Era Purwanto, M.Eng., 
2. Prof. Dr. Mike Yuliana, S.T., M.T., 
3. Prof. Dr. Arman Jaya, S.T., M.T., 
4. Prof. Mochammad Zen Samsono Hadi, S.T., M.Sc., Ph.D., serta
5. Prof. Dr.Eng. Indra Adji Sulistijono, S.T., M.Eng.

Output:
['Arman Jaya', 'Era Purwanto', 'Indra Adji Sulistijono', 'Mike Yuliana', 'Mochammad Zen Samsono Hadi']

```python
pattern = r'(Prof|Dr\.Eng|Dr|Ir)\.\s([A-Za-z]+(\s[A-Za-z]+)+),'

professors = re.findall(pattern, text)
print(professors)

professors = [professor[1] for professor in professors]
print(sorted(professors))
```

### Analisis

1. Karena seluruh nama profesor terdapat di baris pertama, maka cukup untuk hanya membaca baris tersebut dengan fungsi *readline()*
2. Penggunaan *findall()* digunakan untuk mencari semua kata yang memenuhi pola, tidak berhenti di satu kata.
3. Pencarian regex yang dihasilkan berupa tuple, digunakan *list comprehension* untuk mengambil kata terkait (nama profesor) pada index 1. 

### Bedah Pola Regex: 
1. **(Prof|Dr\.Eng|Dr|Ir)** : pengelompokan gelar dengan **| (vertical bar)** sebagai operator OR
2. **\.\s** : setelah gelar terdapat karakter titik dan spasi
3. **([A-Za-z]+(\s[A-Za-z]+)+),** : [A-Za-z]+ menemukan nama pertama, (\s[A-Za-z]+)+ iterasi grup sebanyak apapun untuk menangkap setiap kata pada nama. Tidak terbatas dengan nama yang memiliki dua kata, atau lima kata, atau berapapun.
